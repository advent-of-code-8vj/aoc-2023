```python
def get_input():
    with open("input.txt", "r") as f:
        lines = f.readlines()
    
    for i in range(len(lines)):
        lines[i] = lines[i].replace('\n', '')
    return lines
    
```

```python
def get_match_count(l1, l2):
    return len([x for x in l2 if x in l1])
```

```python
lines = get_input()
lines = list(map(lambda x: x.replace('  ', ' ').replace('  ', ' ').replace('  ', ' ').replace('Card ', '').replace('| ', '|').replace(' |', '|'), lines))
lines = list(map(lambda x: x.split(': '), lines))
lines = list(map(lambda x: (x[0], x[1].split('|')), lines))
lines = list(map(lambda x: (x[0], x[1][0].split(' '), x[1][1].split(' ')), lines))
lines
```

```python
sum = 0
for line in lines:
    matches = get_match_count(line[2], line[1])
    if matches > 0:
        sum = sum + pow(2, matches - 1)
sum
    
```

## Part 2

```python
cards = {}
for line in lines:
    matches = get_match_count(line[2], line[1])
    cards[int(line[0])] = matches
cards
```

```python
def get_won_cards(card):
    sum = 1
    if cards[card] == 0 or card + 1 not in cards:
        return sum
    for i in range(1, cards[card]+1):
        sum = sum + get_won_cards(card + i)
    return sum
    
```

```python
sum = 0
for card in cards:
    sum = sum + get_won_cards(card)
sum
```
